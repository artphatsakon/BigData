```python
# รีสตาร์ท runtime หลังติดตั้ง numpy ก่อนรันโค้ดด้านล่างนี้

# นำเข้าไลบรารี
import pandas as pd
import matplotlib.pyplot as plt
from pmdarima.arima import auto_arima, ADFTest

# 1. Import data
df = pd.read_csv('year_sales.csv')

# 2. Convert Year to datetime
df['Year'] = pd.to_datetime(df['Year'], format='%Y-%m')

# 3. Set Year as index
df.set_index('Year', inplace=True)

# Show plot
df.plot()
plt.title('Yearly Sales')
plt.ylabel('Sales')
plt.show()

# 4. ADF Test using pmdarima's ADFTest
adf_test = ADFTest(alpha=0.05)
should_diff, pval = adf_test.should_diff(df['Sales'])
print(f"Should difference? {should_diff} (p-value={pval})")

# 6. Create train and test datasets
train = df.iloc[:-12]  # last 12 for test
test = df.iloc[-12:]

# 8. Create ARIMA model using auto_arima
model = auto_arima(train['Sales'], seasonal=False, stepwise=True, trace=True)
print(model.summary())

# Forecast
forecast = model.predict(n_periods=len(test))
plt.plot(test.index, test['Sales'], label='Actual')
plt.plot(test.index, forecast, label='Forecast')
plt.title('ARIMA Forecast vs Actual')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.legend()
plt.show()
```
